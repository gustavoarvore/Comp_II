## Bibliotecas

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Função Cúbica

```python
fig, ax = plt.subplots()

x = np.linspace(-10, 10, 100)
y = x**3

plt.plot(x, y)

plt.title('Função Cúbica: y = x^3')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)

plt.show()
```

## 2. Função Modular

```python
fig, ax = plt.subplots()

x = np.linspace(-10, 10, 100)
y = np.abs(x)

plt.plot(x, y)

plt.title('Função Modular: y = |x|')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)

plt.show()
```

## 3. Comparando Três Funções

```python
fig, ax = plt.subplots()

x = np.linspace(-5, 5, 100)

y1 = x
y2 = x**2
y3 = x**3

plt.plot(x, y1, label='y = x')
plt.plot(x, y2, label='y = x^2')
plt.plot(x, y3, label='y = x^3')

plt.title('Comparação de Funções: y = x, y = x^2, y = x^3')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```

## 4. Funções Simétricas

```python
fig, ax = plt.subplots()

x = np.linspace(-10, 10, 100)

y1 = x**2
y2 = -x**2

plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = -x^2')

plt.title('Funções Simétricas: y = x^2 e y = -x^2')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```
